```python
from Assignment.Point import Point
from Assignment.BruteCollinearPoints import BruteCollinearPoints
from Assignment.FastCollinearPoints import FastCollinearPoints
import matplotlib.pyplot as plt


# Create some points
p0 = Point(0, 0)
p1 = Point(1, 1)
p2 = Point(2, 2)
p3 = Point(2, 0)
p4 = Point(3, 3)
p5 = Point(0, 0)

# Draw points
plt.figure(figsize=(8, 6))
plt.title("Point Drawing Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

for point in [p0, p1, p2, p3, p4]:
    point.draw()


# colinear_points = BruteCollinearPoints([p0, p1, p2, p3, p4])
colinear_points = FastCollinearPoints([p0, p1, p2, p3, p4])

segments = colinear_points.segments()
number_of_segments = colinear_points.number_of_segments()

for segment in segments:
    segment.draw()

print(number_of_segments)
```
